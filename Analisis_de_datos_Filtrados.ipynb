```python
#Diccionari: Estructura indice => Valor
Data={
    "Nombre" : ["Suko", "Hamburgueso", "Wilson", "Ellie", "Pepinillo"],
    "Edad" : [22, 23, 24, 25, 26],
    "Ciudad" : ["Cali", "Bogota", "Medellin", "Barranquilla", "Cartagena"]
    }
Data, type(Data)
```

```python
#Data frame: es un arreglo bidimencional que contiene la estructura indice => valor los cuales pueden
#interpretarse como elemetons Nombre dde la columna => Contenido de la clumna
Data_Frame = pd.DataFrame(Data)
Data_Frame
```

```python
#Exportar un data frame
Data_Frame.to_csv("Data_Frame.csv")
```

```python
#Import un csv a data Frame
Data_Frame_2 = pd.read_csv("Data_Frame.csv", index_col = 0)
Data_Frame_2
```

```python
#Selecionar una columna
Nombres = Data_Frame["Nombre"]
print (Nombres)
```

```python
#Seleccionar 1 o mas columnas
Data_Frame[["Nombre", "Edad"]]
```

```python
#Filtrar por indice
Fila = Data_Frame.iloc[0]
Fila
```

```python
#Filtrar por condicin personas mayores de 23 aos
Data_Frame[Data_Frame["Edad"] > 23]
```

```python
#Filtrar por condicin personas menores de 23 aos
Data_Frame[Data_Frame["Edad"] < 23]
```

```python
#Filtrar por condicin personas de 23 aos
Data_Frame[Data_Frame["Edad"] == 23]
```

```python
#Filtrar a personas que son mayores de 23 y aquellos con nombre que inicie con S
Data_Frame[(Data_Frame["Edad"] > 23) & (Data_Frame["Nombre"].str.startswith("E"))]
```

```python
#Filtrar poir query
Data_Frame.query("Edad > 23")
```

```python
#Seleccionar pot Lista "Esta en"
Data_Frame[Data_Frame["Nombre"].isin(["Suko", "Wilson", "Bizcocho"])]
```

```python
#Filtrar por funciones especificas
def Longitud_6 (Nombre):
    return len(Nombre) == 6

Data_Frame[Data_Frame["Nombre"].apply(Longitud_6)]
```

```python
#Filtrar por condiciones multiples
Data_Frame[(Data_Frame["Edad"].between(23,25)) & (Data_Frame["Ciudad"].str.startswith("B"))]
```
